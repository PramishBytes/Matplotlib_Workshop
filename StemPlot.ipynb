```python
import matplotlib.pyplot as plt

x = [10,20,30,40,50]
y= [20,30,40,50,10]

plt.stem(x,y)
plt.show()
```

```python
x = [10,20,30,40,50]
y= [20,30,40,50,10]

plt.stem(x,y,linefmt = ":")
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [10,20,30,40,50]
y= [20,30,40,50,10]

plt.stem(x,y,linefmt = ":",markerfmt = "r+")
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [10,20,30,40,50]
y= [20,30,40,50,10]

plt.stem(x,y,linefmt=":",markerfmt = "ro",bottom = 10)
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [10,20,30,40,50]
y= [20,30,40,50,10]

plt.stem(x,y,linefmt=":",markerfmt = "ro",bottom = -10)
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [10,20,30,40,50]
y= [20,30,40,50,10]

plt.stem(x,y,linefmt=":",markerfmt = "ro",bottom = 10,basefmt="g")
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [10,20,30,40,50]
y= [20,30,40,50,10]

plt.stem(x,y,linefmt=":",markerfmt = "ro",bottom = 10,basefmt = "g",label = "python")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [10,20,30,40,50]
y= [20,30,40,50,10]

plt.stem(x,y,linefmt=":",markerfmt = "ro",bottom = 10,label = "python",orientation="horizontal")
plt.legend()
plt.show()
```
