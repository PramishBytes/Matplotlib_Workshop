## Scatter Plot 

__Syntax__:

import matplotlib.pyplot as plt

x = []

y=[]

plt.scatter(x,y)

plt.show()



__matplotlib.pyplot.scatter__

__matplotlib.pyplot.scatter(x, y, s=None, c=None, *, marker=None, cmap=None, norm=None, vmin=None, vmax=None, alpha=None, linewidths=None, edgecolors=None, colorizer=None, plotnonfinite=False, data=None, **kwargs)__

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5,6,7,8,9]
y = [1,2,3,5,2,3,6,8,9]

plt.scatter(x,y)
plt.show()
```

```python
day = [1,2,3,4,5,6,7,8,9]
no = [1,2,3,5,2,3,6,8,9]
plt.scatter(day,no,color = 'r')
plt.title("scatter plot",fontsize = 15)
plt.xlabel("Day",fontsize = 15)
plt.ylabel("No",fontsize = 15)
plt.show()
```

```python
# add different colors

day = [1,2,3,4,5,6,7,8,9]
no = [1,2,3,5,2,3,6,8,9]
colors = ['r','g','b','y','r','b','g','y','r']
plt.scatter(day,no,c=colors)
plt.title("scatter plot",fontsize = 15)
plt.xlabel("Day",fontsize = 15)
plt.ylabel("No",fontsize = 15)
plt.show()
```

```python
# different sizes

day = [1,2,3,4,5,6,7,8,9]
no = [1,2,3,5,2,3,6,8,9]
colors = ['r','g','b','y','r','b','g','y','r']
sizes = [10,2,4,5,22,15,27,36,18]
plt.scatter(day,no,c=colors,s=sizes)
plt.title("scatter plot",fontsize = 15)
plt.xlabel("Day",fontsize = 15)
plt.ylabel("No",fontsize = 15)
plt.show()
```

```python
# add transparancy using alpha

day = [1,2,3,4,5,6,7,8,9]
no = [1,2,3,5,2,3,6,8,9]
colors = ['r','g','b','y','r','b','g','y','r']
sizes = [10,82,4,5,22,15,27,36,18]
plt.scatter(day,no,c=colors,s=sizes,alpha = 0.2)
plt.title("scatter plot",fontsize = 15)
plt.xlabel("Day",fontsize = 15)
plt.ylabel("No",fontsize = 15)
plt.show()
```

```python
# use of marker parameter

day = [1,2,3,4,5,6,7,8,9]
no = [1,2,3,5,2,3,6,8,9]
colors = ['r','g','b','y','r','b','g','y','r']
sizes = [10,82,4,5,22,15,27,36,18]
plt.scatter(day,no,c=colors,s=sizes,marker = "+")
plt.title("scatter plot",fontsize = 15)
plt.xlabel("Day",fontsize = 15)
plt.ylabel("No",fontsize = 15)
plt.show()
```

```python
#edge color with marker

day = [1,2,3,4,5,6,7,8,9]
no = [1,2,3,5,2,3,6,8,9]
colors = ['r','g','b','y','r','b','g','y','r']
sizes = [10,82,44,35,22,15,27,36,18]
plt.scatter(day,no,c=colors,s=sizes,marker = "*",edgecolor = 'g',linewidth =2)
plt.title("scatter plot",fontsize = 15)
plt.xlabel("Day",fontsize = 15)
plt.ylabel("No",fontsize = 15)
plt.show()
```

```python
# cmap parameter -> color map 

day = [1,2,3,4,5,6,7,8,9]
no = [1,2,3,5,2,3,6,8,9]
colors = [10,20,30,400,120,11.2,40,400,30]
sizes = [100,820,400,500,220,150,270,360,180]
plt.scatter(day,no,c=colors,s=sizes,cmap ="viridis")
plt.colorbar()
plt.title("scatter plot",fontsize = 15)
plt.xlabel("Day",fontsize = 15)
plt.ylabel("No",fontsize = 15)
plt.show()
```

```python
day = [1,2,3,4,5,6,7,8,9]
no = [1,2,3,5,2,3,6,8,9]
colors = [10,20,30,400,120,11.2,40,400,30]
sizes = [100,820,400,500,220,150,270,360,180]
plt.scatter(day,no,c=colors,s=sizes,cmap ="BrBG")
plt.colorbar()
plt.title("scatter plot",fontsize = 15)
plt.xlabel("Day",fontsize = 15)
plt.ylabel("No",fontsize = 15)
plt.show()
```

```python
#plot scatter using multiple parameters as no and no1

day = [1,2,3,4,5,6,7,8,9]
no = [1,2,3,5,2,3,6,8,9]
no1 = [11,22,31,15,2,3,6,8,9]
colors = [10,20,30,400,120,11.2,40,400,30]
sizes = [100,820,400,500,220,150,270,360,180]
plt.scatter(day,no,c=colors,s=sizes,cmap ="BrBG")
plt.scatter(day,no1,c=colors,s=sizes,alpha =0.5)
plt.colorbar()
plt.title("scatter plot",fontsize = 15)
plt.xlabel("Day",fontsize = 15)
plt.ylabel("No",fontsize = 15)
plt.show()
```
