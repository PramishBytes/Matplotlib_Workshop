## MatPlotlib Pie Plot 

import matplotlib.pyplot as plt 

x = []

```python
import matplotlib.pyplot as plt 

x = [10,20,30,40]
y=["c","c++","java","python"]

plt.pie(x)
plt.show()
```

```python
x = [10,20,30,40]
y=["c","c++","java","python"]

plt.pie(x,labels = y)

plt.show()
```

```python
x = [10,20,30,40]
y=["c","c++","java","python"]
ex =[0.4,0.0,0.0,0.0]
plt.pie(x,labels = y,explode = ex)

plt.show()
```

```python
x = [10,20,30,40]
y=["c","c++","java","python"]
ex =[0.2,0.2,0.2,0.2]
plt.pie(x,labels = y,explode = ex)

plt.show()
```

```python
x = [10,20,30,40]
y=["c","c++","java","python"]
ex =[0.2,0.2,0.2,0.2]
c = ["r","b","y","g"]
plt.pie(x,labels = y,explode = ex,colors =c)

plt.show()
```

```python
x = [10,20,30,40]
y=["c","c++","java","python"]
ex =[0.2,0.2,0.2,0.2]
plt.pie(x,labels = y,explode = ex,autopct = "%0.2f%%")

plt.show()
```

```python
x = [10,20,30,40]
y=["c","c++","java","python"]
ex =[0.2,0.2,0.2,0.2]
plt.pie(x,labels = y,explode = ex,autopct = "%0.2f%%",shadow = True)

plt.show()
```

```python
x = [10,20,30,40]
y=["c","c++","java","python"]
ex =[0.2,0.2,0.2,0.2]
plt.pie(x,labels = y,explode = ex,autopct = "%0.2f%%",shadow = True,radius = 1.5)

plt.show()
```

```python
x = [10,20,30,40]
y=["c","c++","java","python"]
ex =[0.2,0.2,0.2,0.2]
plt.pie(x,labels = y,explode = ex,autopct = "%0.2f%%",shadow = True,radius = 1.5,labeldistance = 0.2)

plt.show()
```

```python
x = [10,20,30,40]
y=["c","c++","java","python"]
ex =[0.2,0.2,0.2,0.2]
plt.pie(x,labels = y,explode = ex,autopct = "%0.2f%%",shadow = True,radius = 1.5,labeldistance = 0.2,startangle = 90)

plt.show()
```

```python
x = [10,20,30,40]
y=["c","c++","java","python"]
ex =[0.2,0.2,0.2,0.2]
plt.pie(x,labels = y,explode = ex,autopct = "%0.2f%%",shadow = True,radius = 1.5,labeldistance = 0.2,startangle = 90,
       textprops = {"fontsize":15})

plt.show()
```

```python
x = [10,20,30,40]
y=["c","c++","java","python"]
ex =[0.2,0.2,0.2,0.2]
plt.pie(x,labels = y,explode = ex,autopct = "%0.2f%%",shadow = True,radius = 1.5,labeldistance = 0.2,startangle =0,
       textprops = {"fontsize":15},counterclock = False,wedgeprops = {"linewidth":5})

plt.show()
```

```python
x = [10,20,30,40]
y=["c","c++","java","python"]
ex =[0.2,0.2,0.2,0.2]
plt.pie(x,labels = y,explode = ex,autopct = "%0.2f%%",shadow = True,radius = 1.5,labeldistance = 0.2,startangle =0,
       textprops = {"fontsize":15},counterclock = False,wedgeprops = {"linewidth":5,"edgecolor":"m","width":1.5})

plt.show()
```

```python
x = [10,20,30,40]
y=["c","c++","java","python"]
ex =[0.2,0.2,0.2,0.2]
plt.pie(x,labels = y,explode = ex,autopct = "%0.2f%%",shadow = True,radius = 1.5,labeldistance = 0.2,startangle =0,
       textprops = {"fontsize":15},counterclock = False,wedgeprops = {"linewidth":5,"edgecolor":"m","width":1.5},rotatelabels = True)

plt.show()
```

```python
x = [10,20,30,40]
y=["c","c++","java","python"]
ex =[0.2,0.2,0.2,0.2]
plt.pie(x,labels = y,explode = ex,autopct = "%0.2f%%",shadow = True,radius = 1,labeldistance = 0.2,startangle =0,
       textprops = {"fontsize":15},counterclock = False,wedgeprops = {"linewidth":5,"edgecolor":"m","width":1})
plt.title("PieChart")
plt.show()
```

```python
x = [10,20,30,40]
y=["c","c++","java","python"]
ex =[0.2,0.2,0.2,0.2]
plt.pie(x,labels = y,explode = ex,autopct = "%0.2f%%",shadow = True,radius = 1,labeldistance = 0.2,startangle =0,
       textprops = {"fontsize":15},counterclock = False,wedgeprops = {"linewidth":5,"edgecolor":"m","width":1})
plt.title("PieChart")
plt.legend(loc =1)
plt.show()
```

```python
plt.pie([1])
plt.show()
```

```python
x = [10,20,30,40]
y=["c","c++","java","python"]
x1 = [40,30,20,10]
c = ["r","b","y","g"]
plt.pie(x,labels = y,radius = 1)
plt.pie(x1,radius = 0.5,colors =c)
plt.title("PieChart")

plt.show()
```

```python
x = [10,20,30,40]
y=["c","c++","java","python"]
x1 = [40,30,20,10]
c = ["r","b","y","g"]
plt.pie(x,labels = y,radius = 0.8)
cr = plt.Circle(xy=(0,0),radius = 0.5,facecolor = "w")
plt.gca().add_artist(cr)
plt.title("PieChart")

plt.show()
```
