## Bar Plot


```python
import matplotlib.pyplot as plt


x = ["python","c","c++","Java"]
y= [85,70,60,82]

plt.bar(x,y)
plt.show()
```

## Syntax 

__plt.bar(x,height,width,bottom,align)__

-> x: The categories(eg: fruits)

-> height: The corresponding values(e.g: sales)

-> widthL The width of the bars (default is 0.8)

-> bottom: The baseline for the bars(default is 0)

-> align: How to align bars('center' or 'edge')



```python
plt.xlabel('language',fontsize = 10)
plt.ylabel('No.',fontsize = 10)
plt.title('graph')
plt.bar(x,y)
plt.show()
```

```python
c = ['y','r','m','b']
plt.bar(x,y,width = 0.2,color = c)
plt.show()
```

```python
plt.bar(x,y,width = 0.4,align = 'edge')b
```

```python
plt.bar(x,y,width =0.4, color = c, edgecolor = 'r',linewidth = 2, linestyle = ":",alpha= 0.4)
plt.show()
```
