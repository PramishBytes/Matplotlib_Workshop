## Bar Plot


```python
import matplotlib.pyplot as plt


x = ["python","c","c++","Java"]
y= [85,70,60,82]

plt.bar(x,y)
plt.show()
```

## Syntax 

__plt.bar(x,height,width,bottom,align)__

-> x: The categories(eg: fruits)

-> height: The corresponding values(e.g: sales)

-> widthL The width of the bars (default is 0.8)

-> bottom: The baseline for the bars(default is 0)

-> align: How to align bars('center' or 'edge')



```python
plt.xlabel('language',fontsize = 10)
plt.ylabel('No.',fontsize = 10)
plt.title('graph')
plt.bar(x,y)
plt.show()
```

```python
c = ['y','r','m','b']
plt.bar(x,y,width = 0.2,color = c)
plt.show()
```

```python
plt.bar(x,y,width = 0.4,align = 'edge')b
```

```python
plt.bar(x,y,width =0.4, color = c, edgecolor = 'r',linewidth = 2, linestyle = ":",alpha= 0.4)
plt.show()
```

```python
plt.bar(x,y,width = 0.4, color = 'r',edgecolor = 'r',label = 'graph')
plt.legend()
```

```python
x = ['python','c','c++','java']
y = [85,70,60,82]
z = [20,30,40,50]
plt.xlabel("language",fontsize = 15)
plt.ylabel("No..",fontsize=15)
plt.title("Graph",fontsize = 15)
c = ["y","b","m","g"]
plt.bar(x,y,width =0.4,color = "r",label = "popularity")
plt.bar(x,z,width =0.4,color = "g",label = "popularity1")
plt.legend()
plt.show()
```

```python
import numpy as np

x = ['python','c','c++','java']
y = [85,70,60,82]
z = [20,30,40,50]

width = 0.4
p = np.arange(len(x))
p1 = [j+width for j in p]

plt.xlabel("language",fontsize = 15)
plt.ylabel("No..",fontsize=15)
plt.title("Graph",fontsize = 15)
c = ["y","b","m","g"]

plt.bar(p,y,width =0.4,color = "r",label = "popularity")
plt.bar(p1,z,width =0.4,color = "g",label = "popularity1")

plt.xticks(p+width/2,x,rotation = 45)
plt.legend()
plt.show()
```

## horizontal 


```python
x = ['python','c','c++','java']
y = [85,70,60,82]
z = [20,30,40,50]
plt.xlabel("language",fontsize = 15)
plt.ylabel("No..",fontsize=15)
plt.title("Graph",fontsize = 15)
c = ["y","b","m","g"]
plt.barh(x,y,color = "r",label = "popularity")
plt.barh(x,z,color = "g",label = "popularity1")
plt.legend()
plt.show()
```
